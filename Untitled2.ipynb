## Data wrangling

### Subtask:
Further wrangle the data in `df_cleaned` to prepare it for analysis and visualization.


**Reasoning**:
Convert data types, engineer new features, and handle remaining inconsistencies in the `df_cleaned` DataFrame.



```python
import pandas as pd

# Convert 'Arrest Month' to categorical
df_cleaned['Arrest Month'] = pd.Categorical(df_cleaned['Arrest Month'], categories=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], ordered=True)

# Feature Engineering: Drug-related offense
df_cleaned['Drug Offense'] = df_cleaned['Offense Statute Grouped'].apply(lambda x: 1 if x == 'Drug-related offenses' else 0)

# Feature Engineering: Year of Arrest (already present, check type)
if not pd.api.types.is_numeric_dtype(df_cleaned['Arrest Year']):
    df_cleaned['Arrest Year'] = pd.to_numeric(df_cleaned['Arrest Year'], errors='coerce')

# Feature Engineering: Age Groups
def categorize_age(age):
    if age is None or pd.isna(age):
        return "Unknown"
    if age < 25:
        return "Young Adult"
    elif age < 65:
        return "Adult"
    else:
        return "Senior"

df_cleaned['Age Group'] = df_cleaned['Age Processed'].apply(categorize_age)

# Handle inconsistencies in categorical columns (example: Race)
race_mapping = {
    'White (incl. Hispanic origin)': 'White',
    'Black (incl. Hispanic origin)': 'Black',
    'Asian (incl. Hispanic origin)': 'Asian',
    # Add more mappings as needed
}

df_cleaned['Race'] = df_cleaned['Race'].map(race_mapping).fillna('Other')

# Display the updated DataFrame
display(df_cleaned.head())
```

## Data analysis

### Subtask:
Analyze the cleaned and wrangled data (df_cleaned) to gain insights.


**Reasoning**:
Calculate descriptive statistics for numerical columns and explore the distribution of categorical features.  Investigate relationships between variables using cross-tabulations and grouped summaries. Calculate key metrics like the frequency of different offense categories and the average age of offenders for each category.



```python
# Descriptive Statistics for Numerical Columns
print("Descriptive Statistics for 'Age Processed':")
display(df_cleaned['Age Processed'].describe())

# Distribution of Categorical Features
categorical_cols = ['Offense Category', 'Offense Statute Grouped', 'Race', 'Gender', 'Age Group']
for col in categorical_cols:
    print(f"\nValue Counts for {col}:")
    display(df_cleaned[col].value_counts())

# Relationship between 'Offense Category' and 'Age Group'
print("\nRelationship between Offense Category and Age Group:")
display(pd.crosstab(df_cleaned['Offense Category'], df_cleaned['Age Group']))

# Relationship between 'Drug Offense' and 'Race'
print("\nRelationship between Drug Offense and Race:")
display(pd.crosstab(df_cleaned['Drug Offense'], df_cleaned['Race']))

# Distribution of 'Age Processed' for different 'Offense Statute Grouped' categories
print("\nDistribution of Age Processed for different Offense Statute Grouped categories:")
display(df_cleaned.groupby('Offense Statute Grouped')['Age Processed'].describe())

# Key Metrics
print("\nFrequency of Different Offense Categories:")
display(df_cleaned['Offense Category'].value_counts(normalize=True))

print("\nAverage Age of Offenders for Each Offense Category:")
display(df_cleaned.groupby('Offense Category')['Age Processed'].mean())

print("\nProportion of Drug-Related Offenses within Each Race Group:")
drug_offenses_by_race = df_cleaned.groupby('Race')['Drug Offense'].value_counts(normalize=True).unstack()
display(drug_offenses_by_race)
```

## Data visualization

### Subtask:
Visualize key findings from the data analysis performed on `df_cleaned`.


**Reasoning**:
Visualize the key findings from the data analysis using various plot types as instructed.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram of 'Age Processed'
plt.figure(figsize=(8, 6))
sns.histplot(df_cleaned['Age Processed'], kde=True)
plt.title('Distribution of Age Processed')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 2. Bar chart of 'Offense Category' frequency
plt.figure(figsize=(12, 6))
offense_category_counts = df_cleaned['Offense Category'].value_counts().sort_values(ascending=False)
sns.barplot(x=offense_category_counts.index, y=offense_category_counts.values)
plt.title('Frequency of Offense Categories')
plt.xlabel('Offense Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 3. Grouped bar chart of 'Offense Category' and 'Age Group'
plt.figure(figsize=(14, 8))
sns.countplot(x='Offense Category', hue='Age Group', data=df_cleaned)
plt.title('Offense Category by Age Group')
plt.xlabel('Offense Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. Bar chart of drug-related offenses by 'Race'
plt.figure(figsize=(10, 6))
drug_offenses_by_race = df_cleaned.groupby('Race')['Drug Offense'].mean()
sns.barplot(x=drug_offenses_by_race.index, y=drug_offenses_by_race.values)
plt.title('Proportion of Drug-Related Offenses by Race')
plt.xlabel('Race')
plt.ylabel('Proportion of Drug Offenses')
plt.show()


# 5. Box plot of 'Age Processed' by 'Offense Statute Grouped'
plt.figure(figsize=(12, 6))
sns.boxplot(x='Offense Statute Grouped', y='Age Processed', data=df_cleaned)
plt.title('Distribution of Age Processed by Offense Statute Grouped')
plt.xlabel('Offense Statute Grouped')
plt.ylabel('Age Processed')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
