## Data cleaning

### Subtask:
Clean the data by handling inconsistencies and converting data types where necessary.


**Reasoning**:
Clean the 'Race' and 'Gender' columns by replacing "Unknown" and "Not Provided" with NaN, and then clean the 'Age Range' column by converting it to numerical representation.  Finally, handle duplicate rows and check for other inconsistencies.



```python
# Replace "Unknown" and "Not Provided" with NaN in 'Race' and 'Gender'
for col in ['Race', 'Gender']:
    df[col] = df[col].replace(['Unknown', 'Not Provided'], pd.NA)

# Clean 'Age Range' column
def clean_age_range(age_range):
    if pd.isna(age_range):
        return pd.NA
    try:
        age_parts = age_range.split('-')
        if len(age_parts) == 2:
            lower_bound = int(age_parts[0])
            upper_bound = int(age_parts[1])
            return (lower_bound + upper_bound) / 2
        else:
            return pd.NA  # Handle cases with invalid format
    except ValueError:
        return pd.NA

df['Cleaned_Age'] = df['Age Range'].apply(clean_age_range)


# Handle duplicate rows (keep first occurrence)
df.drop_duplicates(inplace=True)

# Check for other inconsistencies (example: Offense Category)
print(df['Offense Category'].value_counts())
# Further investigation into other columns may be needed
# Example: unusual values, typos, or formatting issues in 'Offense Statute'
print(df['Offense Statute'].value_counts())

display(df.head())
```
