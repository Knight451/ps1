Merge files and analyze data for visualization.

## Data loading

### Subtask:
Load the three Excel files into pandas DataFrames.


**Reasoning**:
Load the three Excel files into pandas DataFrames as specified in the instructions.



```python
import pandas as pd

try:
    df_2023 = pd.read_excel("Criminal Justice Data Dashboard - defendant charges 2023 (1).xlsx")
    df_2021 = pd.read_excel("Criminal Justice Data Dashboard - defendant charges 2021.xlsx")
    df_2022 = pd.read_excel("Criminal Justice Data Dashboard - defendant charges 2022.xlsx")
    print("DataFrames loaded successfully.")
except FileNotFoundError:
    print("One or more files not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Data preparation

### Subtask:
Prepare the data for merging by ensuring consistency in column names and data types across the three dataframes: `df_2021`, `df_2022`, and `df_2023.


**Reasoning**:
Standardize column names to lowercase with underscores, ensure data type consistency, and handle the extra column in df_2022.



```python
import pandas as pd

def standardize_column_names(df):
    df.columns = [col.lower().replace(' ', '_') for col in df.columns]
    return df

df_2021 = standardize_column_names(df_2021)
df_2022 = standardize_column_names(df_2022)
df_2023 = standardize_column_names(df_2023)

# Handle the extra column in df_2022
if 'unnamed:_11' in df_2022.columns:
    df_2022 = df_2022.drop(columns=['unnamed:_11'])

# Ensure consistency in 'arrest_year'
for df in [df_2021, df_2022, df_2023]:
    df['arrest_year'] = pd.to_numeric(df['arrest_year'], errors='coerce')

# Display the first few rows of each dataframe to inspect
display(df_2021.head())
display(df_2022.head())
display(df_2023.head())
```

## Data wrangling

### Subtask:
Merge the three DataFrames (`df_2021`, `df_2022`, and `df_2023`) into a single DataFrame named `df_merged`.


**Reasoning**:
Concatenate the three dataframes vertically and reset the index.



```python
df_merged = pd.concat([df_2021, df_2022, df_2023], ignore_index=True)
display(df_merged.head())
```

## Data exploration

### Subtask:
Explore the merged DataFrame (`df_merged`) to identify potential issues like missing values, outliers, and inconsistencies. Calculate descriptive statistics for numerical columns and analyze the distribution of key variables.


**Reasoning**:
Check for missing values, calculate descriptive statistics for numerical columns, and analyze the distribution of categorical variables in `df_merged`.



```python
# Check for missing values
missing_values = df_merged.isnull().sum()
print("Missing Values:\n", missing_values)

# Descriptive statistics for numerical columns
numerical_cols = df_merged.select_dtypes(include=['number']).columns
descriptive_stats = df_merged[numerical_cols].describe()
print("\nDescriptive Statistics:\n", descriptive_stats)

# Analyze distribution of categorical variables
categorical_cols = df_merged.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nDistribution of {col}:\n", df_merged[col].value_counts())
```

## Data analysis

### Subtask:
Analyze trends and patterns in defendant charges across the three years (2021-2023).


**Reasoning**:
Calculate the total number of charges for each year and determine the most frequent charge types for each year and overall. Analyze changes in charge frequency over time.



```python
# Calculate total charges per year
yearly_charges = df_merged.groupby('arrest_year').size()
print("Total charges per year:\n", yearly_charges)

# Determine most frequent charge types per year and overall
for year in yearly_charges.index:
    yearly_df = df_merged[df_merged['arrest_year'] == year]
    top_offense_category = yearly_df['offense_category'].value_counts().idxmax()
    top_offense_statute = yearly_df['offense_statute'].value_counts().idxmax()
    print(f"\nMost frequent charge types for {year}:")
    print(f"Offense Category: {top_offense_category}")
    print(f"Offense Statute: {top_offense_statute}")

top_overall_offense_category = df_merged['offense_category'].value_counts().idxmax()
top_overall_offense_statute = df_merged['offense_statute'].value_counts().idxmax()

print(f"\nMost frequent charge types overall:")
print(f"Offense Category: {top_overall_offense_category}")
print(f"Offense Statute: {top_overall_offense_statute}")

# Analyze changes in charge frequency over time (yearly proportions)
yearly_proportions = df_merged.groupby('arrest_year')['offense_category'].value_counts(normalize=True)
print("\nYearly proportions of offense categories:\n", yearly_proportions)
```

**Reasoning**:
Group and aggregate data by relevant categories (race, age\_range, gender) to calculate the frequency of charges within each group and across years. Look for disparities in charge types across demographic groups.



```python
# Group and aggregate data by demographic categories
for category in ['race', 'age_range', 'gender']:
    grouped_data = df_merged.groupby([category, 'arrest_year'])['offense_category'].value_counts(normalize=True)
    print(f"\nProportions of offense categories by {category} and year:\n", grouped_data)

# Further analysis for disparities
# Example: Analyze disparities in drug charges across races
drug_charges_by_race = df_merged[df_merged['offense_category'] == 'Drugs'].groupby(['race', 'arrest_year']).size()
print(f"\nNumber of drug charges by race and year:\n{drug_charges_by_race}")
```

**Reasoning**:
Identify notable trends or patterns, such as increases or decreases in certain crime types, changes in demographics of individuals charged, etc.  Based on the previous analysis, I will focus on visualizing the trends in drug charges across different demographics.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize drug charges by race and year
plt.figure(figsize=(10, 6))
drug_charges_by_race.unstack().plot(kind='bar')
plt.title('Drug Charges by Race and Year')
plt.xlabel('Race')
plt.ylabel('Number of Charges')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()

# Visualize proportions of offense categories by age range and year
plt.figure(figsize=(12, 6))
grouped_data = df_merged.groupby(['age_range', 'arrest_year'])['offense_category'].value_counts(normalize=True)
grouped_data = grouped_data.unstack()
grouped_data.plot(kind='bar', stacked=True)
plt.title('Proportions of Offense Categories by Age Range and Year')
plt.xlabel('Age Range')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Offense Category')
plt.tight_layout()
plt.show()

# Visualize proportions of offense categories by gender and year
plt.figure(figsize=(8, 6))
grouped_data = df_merged.groupby(['gender', 'arrest_year'])['offense_category'].value_counts(normalize=True)
grouped_data = grouped_data.unstack()
grouped_data.plot(kind='bar', stacked=True)
plt.title('Proportions of Offense Categories by Gender and Year')
plt.xlabel('Gender')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Offense Category')
plt.tight_layout()
plt.show()
```
